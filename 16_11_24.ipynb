```python
from zeep import Client

# WSDL'den Client oluştur
client = Client('http://webservices.oorsprong.org/websamples.countryinfo/CountryInfoService.wso?WSDL')

# Tür tanımlamasını al
country_code_type = client.get_type('ns0:tCountryCodeAndName')

# Türü kullanarak veri oluştur
country_request = country_code_type(sISOCode='US', sName='United States')

# Servis çağrısı
response = client.service.CapitalCity(sCountryISOCode=country_request.sISOCode)

print(f"Başkent: {response}")
```

```python
from zeep import Client
from zeep.transports import Transport
import pandas as pd
transport  = Transport(timeout=10)

client = Client('http://webservices.oorsprong.org/websamples.countryinfo/CountryInfoService.wso?WSDL',transport=transport)
client.settings(raw_response = True)



countries = client.service.FullCountryInfoAllCountries()

# country_list = [
#     {
#         'name':country['sName'],
#         'languages':','.join( [l['sName'] for l in country['Languages']['tLanguage']])
#     }
#     for country in countries if country['Languages']
# ]

country_list = [
    {
        'name': country['sName'],
        'languages': ','.join([l['sName'] for l in country['Languages']['tLanguage'] if l != ''])
    }
    for country in countries if country['Languages']
]

df = pd.DataFrame(country_list)
df.sort_values(by=['name'], inplace=True)
df.to_excel('./test.xlsx',index=False)
```
