```python
import requests
import xml.etree.ElementTree as ET

url = "https://tckimlik.nvi.gov.tr/Service/KPSPublic.asmx"

soap_body = f"""
<soapenv:Envelope xmlns:soapenv="http://schemas.xmlsoap.org/soap/envelope/" xmlns:kps="http://tckimlik.nvi.gov.tr/WS">
   <soapenv:Header/>
   <soapenv:Body>
      <kps:TCKimlikNoDogrula>
         <kps:TCKimlikNo>12345678901</kps:TCKimlikNo>
         <kps:Ad>ADI</kps:Ad>
         <kps:Soyad>SOYADI</kps:Soyad>
         <kps:DogumYili>1980</kps:DogumYili>
      </kps:TCKimlikNoDogrula>
   </soapenv:Body>
</soapenv:Envelope>
"""
headers = {
    "Content-Type": "text/xml; charset=utf-8",
    "SOAPAction": "http://tckimlik.nvi.gov.tr/WS/TCKimlikNoDogrula"
}

response = requests.post(url, data=soap_body, headers=headers)
root = ET.fromstring(response.content)
body = root.find('{http://schemas.xmlsoap.org/soap/envelope/}Body')
response = body.find('{http://tckimlik.nvi.gov.tr/WS}TCKimlikNoDogrulaResponse')
result = response.find('{http://tckimlik.nvi.gov.tr/WS}TCKimlikNoDogrulaResult')
result.text
```

```python
import pandas as pd
# ulke bilgilerini getirir ve excel dosyasına kaydeder.

wsdl= 'http://webservices.oorsprong.org/websamples.countryinfo/CountryInfoService.wso?WSDL'
client = Client(wsdl)
data = None
with client.settings(raw_response=True): # cevabı direk xml olarak almak icin
    data = client.service.FullCountryInfoAllCountries().text

# namespaceler
soap_ns = '{http://www.w3.org/2003/05/soap-envelope}'
m_ns = '{http://www.oorsprong.org/websamples.countryinfo}'
root = ET.fromstring(data)

# for lan in root.findall(f'.//{m_ns}tLanguage'):
#     code = lan.find(f'{m_ns}sISOCode').text
#     iso_codes.append(code)

countries =[]
for country in root.findall(f'.//{m_ns}tCountryInfo'):
    name = country.find(f'{m_ns}sName').text
    phone_code = country.find(f'{m_ns}sPhoneCode').text
    flag_url = country.find(f'{m_ns}sCountryFlag').text
    languages = [(k.find(f'{m_ns}sName').text) for k in country.findall(f'.//{m_ns}tLanguage')]
    countries.append({'name':name,'phone_code':phone_code,'flag_url':flag_url, 'languages':','.join(languages)})
df= pd.DataFrame(countries)

for index,row in df.iterrows():
    if row['languages'] == '':
        df.loc[index,'languages'] = '-'
df.sort_values(by=['name'], inplace=True)

df.to_excel('./countries.xlsx',index=False)
df
```
